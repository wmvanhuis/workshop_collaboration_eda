```python
# Correlation analysis with target variable
correlations = df.corr()["target"].sort_values(ascending=False)
print("\nTop correlated features with target:")
print(correlations.head(10))
```

```python
# Select the top 5 features most correlated with target
selected_features = correlations.index[1:6]  # Skipping "target" itself
selected_columns = ["target"] + list(selected_features)
```

```python
# Create the processed dataset
processed_df = df[selected_columns]
```

```python
# Save the processed dataset
processed_data_path = 'data/processed/processed_data.csv'
processed_df.to_csv(processed_data_path, index=False)
```
